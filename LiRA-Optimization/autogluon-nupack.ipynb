```python
import pandas as pd
import numpy as np
df = pd.read_csv('/kaggle/input/toehold-optimization/final.csv')
df.head()
```

```python
# get columns that has variation of complexes
cols_complex = []
for i in df.columns[df.columns.str.startswith('on_level_')].tolist()[1:-1]:
    cols_complex.append(i.split('on_level_')[1])

# get the list of columns to get into regression
_param = ['_mfe',
          '_pfunc',
          '_free_energy',
          '_ensemble_size',
          '_mfe_energy',
          '_mfe_stack_energy',
          '_t1',
          '_t1_rank']
param_  = ['on_level_',
            'off_level_',
            'on_mean_',
            'on_sum_',
            'on_stdev_',
            'off_mean_',
            'off_sum_',
            'off_stdev_']
param = ['mfe_ABS-miR21bottomseal1', 'mfe_toeholdmir21-miR92topseal',
          'mfe_miR92bottomseal2-end',
          'gc_ABS-miR21bottomseal1', 'gc_toeholdmir21-miR92topseal',
          'gc_miR92bottomseal2-end',
          'gc_content', 'structure_prob', 'defect']

def get_param(df, _param, cols_complex):

  params_all = []
  for i in cols_complex:
    for j in _param:
      params_all.append(str(i+j))
    # for k in param_:
    #   params_all.append(str(k+i))
  params_all.extend(param)
  return(params_all)
```

```python
# Check which rows contain infinity values
rows_with_inf = df.isin([np.inf]).any(axis=1)
print(rows_with_inf)
# Drop rows with infinity values
df_non_inf = df.drop(index = pd.Series(rows_with_inf)[pd.Series(rows_with_inf)].index[0])
params_all = get_param(df, _param, cols_complex)
print(params_all)
```

```python
def train_auto(label, df, params_all):
    
    from autogluon.tabular import TabularDataset, TabularPredictor
    train_data = TabularDataset(df[params_all])
    # df_non_inf
    # Check which rows contain infinity values
    rows_with_inf = df[params_all].isin([np.inf]).any(axis=1)
    print(rows_with_inf)

    save_path = './model_ag/'+ label
    predictor = TabularPredictor(label=label, path=save_path, eval_metric="r2").fit(train_data)
    fi = predictor.feature_importance(train_data)
    fi.to_csv(save_path+'/fi.csv')
    print(label)
    display(fi)
    predictor = TabularPredictor.load("/kaggle/working/"+save_path+'/')
    print(predictor.info())
    ld = predictor.leaderboard(extra_info=True, silent = True)
    ld.to_csv(save_path+'/leaderboard.csv')
    display(ld)
```

```python
metrics_inf = ['on_off_ratio_avg', 'weighted_on_off_ratio_avg', 'on_off_ratio_(a+c+b)']
metrics_non = ['on_off_minus_avg','weighted_on_off_minus_avg', 'on_level_avg', 'off_level_avg', 
           'weighted_on_level_avg', 'weighted_off_level_avg']
# print(metrics)
for i in metrics_inf:
    print(i)
    print(df_non_inf[i].unique())
for i in metrics_non:
    print(i)
    print(df[i].unique())
```

```python
# label = 'weighted_on_off_minus_avg'
for label in metrics_inf:
    params_all.append(label)
    print(params_all)
    train_auto(label, df_non_inf, params_all)
    params_all.remove(label)
    
for label in metrics_non:
    params_all.append(label)
    print(params_all)
    train_auto(label, df, params_all)
    params_all.remove(label)
```
